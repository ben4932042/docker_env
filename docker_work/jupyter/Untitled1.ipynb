```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import lxml
for page in range(41):
    url = 'https://zan01.com/lotto649/history/{}'.format(page+1)

    res = requests.get(url=url)
    soup = BeautifulSoup(res.text,'html.parser')
    table = soup.select('html[lang="zh-tw"]')
    content = table[0].select('tbody')
    a = content[0].select('tr')
    c = a[0].text.replace(' ','').replace('\n','').split('\r')
    with open('./lo.txt','a') as outfile:
        for i in range(len(c)):
            try:
                if int(c[i]) < 109000056 and int(c[i]) > 104000001 :
                        outfile.write(str(c[i:i+4])+'|')
            except:
                pass
    print('page{} success!'.format(page+1))
```

```python
with open('./lo.txt','r') as outfile:
    file = outfile.readline()
```

```python
file_list = file.split('|')
```

```python
L = []
for data in file_list:
    L.append([data[2:11],data[15:25],data[29:41],data[45:47]])
del L[-1]
```

```python
df = pd.DataFrame(L)
df[4],df[5],df[6],df[7],df[8],df[9] ='','','','','',''
```

```python
for i in range(len(df)):
    df[4][i] = df[2][i][0:2]
    df[5][i] = df[2][i][2:4]
    df[6][i] = df[2][i][4:6]
    df[7][i] = df[2][i][6:8]
    df[8][i] = df[2][i][8:10]
    df[9][i] = df[2][i][10:12]
```

```python
df[10] = df[3]
df = df.drop([2,3],axis = 1)
```

```python
df = df.rename(columns={0:'index' , 1: 'Date',4:'num1',5:'num2',6:'num3',7:'num4',8:'num5',9:'num6',10:'num7'})
df.index = pd.to_datetime(df['Date'],format = '%Y/%m/%d')
```
