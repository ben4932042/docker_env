```python
def choose_legel_person(stock='2330',broker='1020',ass_date = '2020-04-19',rolling_time = 20):
    
    df = pd.read_json(r'./{stock}/{broker}.json'.format(stock=stock ,broker=broker))\
    .T.drop(0,axis = 0).drop(labels=[3,4],axis='columns')\
        .reset_index(drop = True).rename(columns={0:'Date' , 1: 'buy',2:'sell'})
    for i in range(len(df['Date'])):
        df['Date'][i] = df['Date'][i].replace('/','-')
    df.index = pd.to_datetime(df['Date'],format = '%Y/%m/%d')

    df['buy'] ,df['sell'] = df['buy'].astype('float64') ,df['sell'].astype('float64')
    
    CSV = './all_stock_price/public/2330.TW.csv'
    daily_trade =pd.read_csv('{}'.format(CSV)).drop(['Dividends','Stock Splits'],axis = 1)
    daily_trade.index = pd.to_datetime(daily_trade['Date'],format = '%Y/%m/%d')
    df['Cost'] = ((daily_trade['High']+daily_trade['Low'])*(0.5))*(df['buy'] - df['sell'])
    avg_buy = df['Cost'][df['Cost']>=0][ass_date:].head(rolling_time).sum()/len(df['Cost'][df['Cost']>=0][ass_date:].head(rolling_time))
    avg_sell = df['Cost'][df['Cost']<0][ass_date:].head(rolling_time).sum()/len(df['Cost'][df['Cost']<0][ass_date:].head(rolling_time))
    
    
    return [stock,broker,avg_buy,avg_sell]
```
