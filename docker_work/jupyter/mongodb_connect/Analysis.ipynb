```python
import time
start = time.time()
for broker in brokers:
    date_list ,value_list = [],[],
    for data in collection.find({'broker':broker}):   
        date_list.append(data['date'])
        value_list.append(data['value'])

    Dic = {'date':date_list,'value':value_list}
    df = pd.DataFrame(Dic)
    df.date = pd.to_datetime(df['date'],format = '%Y/%m/%d')
    df2 = df[df['date']>'2020-01-01'].set_index('date')
    df3 = pd.concat([df1,df2],axis=1)
    df4 = df3.dropna()
    df4['price'] = df4['price'].astype('float')
    df4['value'] = df4['value'].astype('float')
    df4['result'] = df4['price']*df4['value']    
    price = df4['result'].sum()
    value = df4['value'].sum()
    Result_Dic[broker] = [-price,value]
    print('{} finish!'.format(broker))
end = time.time()

print(end - start)
```
