```python
import pandas as pd
import datetime


def create_assist_date(datestart=None, dateend=None):
    if datestart is None:
#         datestart = '2016-01-01'
        datestart = '2020-06-04'
    if dateend is None:
#         dateend = datetime.datetime.now().strftime( '%Y-%m-%d' )
# test
        dateend = datetime.datetime.now().strftime( '%Y-%m-%d' )


    datestart = datetime.datetime.strptime( datestart, '%Y-%m-%d' )
    dateend = datetime.datetime.strptime( dateend, '%Y-%m-%d' )
    date_list = []
    date_list.append( datestart.strftime( '%Y/%m/%d' ) )
    while datestart < dateend:
        datestart += datetime.timedelta( days=+1 )
        date_list.append( datestart.strftime( '%Y/%m/%d' ) )
    return date_list

Date_list = create_assist_date()
for times in Date_list:
    time = times.split('/')
    times = times.replace('/','-')
    Del_list = []
    url = 'https://www.tpex.org.tw/web/stock/aftertrading/otc_quotes_no1430/stk_wn1430_result.php?l=zh-tw&o=json&d={Year}/{Month}/{Day}&se=EW&s=0,asc,0'.format(Year=int(time[0]) - 1911 ,Month=time[1] ,Day=time[2])
    df = pd.read_json(url)
    df = df.drop(['reportDate','iTotalRecords','colNum'],axis = 1)
    df['stockiid'] = ''
    df['date'] = times
    df['open'] = ''
    df['high'] = ''
    df['low'] = ''
    df['close'] = ''
    df['volume'] = ''
    for index in range(len(df['aaData'])):
        data = df['aaData'][index]
        df['stockiid'][index] = data[0]
        df['open'][index] = data[4]
        df['high'][index] = data[5]
        df['low'][index] = data[6]
        df['close'][index] = data[2]
        df['volume'][index] = data[9]
    df = df.drop(['aaData'],axis = 1)
    for i in range(len(df)):
        if len(df['stockiid'][i]) > 4:
            Del_list.append(i)
    df = df.drop(Del_list,axis = 0).reset_index(drop = True)
    
```
