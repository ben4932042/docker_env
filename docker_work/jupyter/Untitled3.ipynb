```python
import pandas as pd
```

```python
df = pd.read_csv('./Project_CSV_raw/stock_tw.csv')
```

```python
df.drop(['出表日期','公司名稱','公司簡稱','外國企業註冊地國','住址','營利事業統一編號','董事長','總經理',
        '發言人','發言人','發言人職稱','代理發言人','總機電話','成立日期','上市日期','普通股每股面額',
        '實收資本額','私募股數','特別股','編制財務報表類型','股票過戶機構','過戶電話','過戶地址',
        ''],axis=1)
```

```python
df = df.T.reset_index(drop=True)
```

```python
df.to_csv('./Project_CSV_result/tuning_score_everyone.csv')
```

```python
df[0]
```

```python
dfr = df
```

```python
Dic = {}
for i in range(50,101):
    I = '{:.2f}'.format(i*0.01)
    Dic[I] = 0

for i in range(len(dfr)):
    k = '{:.2f}'.format(dfr['auc'][i])
    if k in Dic.keys():
        Dic[k] = Dic[k]+1
    else:
        Dic[k] = 1

result1 = pd.Series(Dic,Dic.keys()).to_frame().reset_index().sort_values(['index']).reset_index(drop=True)
```
