```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./test_locate.csv')
df = df.drop(['Unnamed: 0'],axis = 1)
```

```python
for i in range(len(df['Latitude_and_Longitude'])):
    if df['Latitude_and_Longitude'][i] =='error':
        df['Latitude_and_Longitude'][i] = np.nan
    else:
        df['Latitude_and_Longitude'][i] = [float((df['Latitude_and_Longitude'][i])[1:-1].replace('\'','').split(',')[0]),float((df['Latitude_and_Longitude'][i])[1:-1].replace('\'','').split(',')[1])]
```
